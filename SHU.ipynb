```python
excelname= "kd"
sheetname = "kd"

wb = openpyxl.load_workbook(str(excelname)+".xlsx")
ws = wb[str(sheetname)]

dataset = []

i = 1
while(ws.cell(row=i, column=1).value!=None):
    row = []
    j = 1
    while(j<=5):
        row.append(ws.cell(row=i, column=j).value)
        j = j +1
    dataset.append(row)
    i = i + 1
```

```python
pd.DataFrame(dataset)
```

```python
trade =[[0 for x in range(11)] for y in range(len(dataset))]
trade[0][0] = "date"
trade[0][1] = "open"
trade[0][2] = "high"
trade[0][3] = "low"
trade[0][4] = "close"
trade[0][5] = "rsv"
trade[0][6] = "k"
trade[0][7] = "d"
trade[0][8] = "buy_day"
trade[0][9] = "cum_day"
trade[0][10] = "sell_day"
```

```python
pd.DataFrame(trade).head()
```

```python
for i in range(1, len(trade)):
    trade[i][0] = dataset[i][0]
    trade[i][1] = dataset[i][1]
    trade[i][2] = dataset[i][2]
    trade[i][3] = dataset[i][3]
    trade[i][4] = float(dataset[i][4])
    
for i in range(9, len(trade)): 
    row=[]
    a=[]
    K=50
    D=50
    for j in range(0, 9):
        row.append(trade[i-j][2])
        max(row)
        a.append(trade[i-j][3])
        min(a)
        
    RSV =((trade[i][4]) - min(a))/(max(row) - min(a))*100
    trade[i][5]=RSV
    K= K*2/3 + RSV*1/3
    D= D*2/3 + K*1/3
    trade[i][6]=K
    trade[i][7]=D
    
 
```
